```python
y = df['target']
X = df.drop(columns="target")
X_design = sm.add_constant(X)
```

`Answer:` Adding a constant is necessary because, without it, you force the model to pass through the origin (0,0). This is usually unrealistic and leads to biased coefficients because the model 'tilts' the line to hit zero instead of fitting the actual data trend.

## Question 6 — Multicollinearity
1. Compute the correlation matrix of the predictors.
2. Identify any strongly correlated variables.

👉 *Explain how multicollinearity affects coefficient estimates.*

```python
#1 Computing the correlation matrix of the predictors
corr_matrix = X.corr()
```

```python
for col in corr_matrix:
  for row in corr_matrix:
    if (corr_matrix[col][row] >= 0.5) and col > row:
      print(col, "and", row, corr_matrix[col][row])
```

`Answer:`The strong correlation between s1 and s2 (r ≈ 0.90) indicates significant multicollinearity, which destabilizes coefficient estimates by inflating their standard errors and making them highly sensitive to small changes in the data. This redundancy prevents the model from isolating the unique effect of each predictor, often resulting in statistically insignificant p-values or logically incorrect signs for the coefficients. Essentially, when predictors are this closely linked, the regression cannot accurately determine which specific variable is driving the response.
